```python
# import library
from tabulate import tabulate

# Membuat Class Transaksi
class Transaksi():
    def __init__(self):
        self.nama_barang = []
        self.harga_barang = []
        self.jumlah_barang = []

# Skenario menambah barang
    def tambah_barang(self, nama, harga, jumlah):
        self.nama_barang.append(nama)
        self.harga_barang.append(harga)
        self.jumlah_barang.append(jumlah)

# Skenario update barang
    def update_barang(self, index, nama=None, harga=None, jumlah=None):
        if nama is not None:
            self.nama_barang[index] = nama
        if harga is not None:
            self.harga_barang[index] = harga
        if jumlah is not None:
            self.jumlah_barang[index] = jumlah
        return "Barang berhasil diupdate"

# Skenario hapus barang
    def hapus_barang(self, index):
        try:
            self.nama_barang.pop(index)
            self.harga_barang.pop(index)
            self.jumlah_barang.pop(index)
            print("Barang berhasil dihapus.")
        except IndexError:
            print("Terjadi kesalahan. Barang tidak dapat dihapus.")

# Skenario hapus semua transaksi
    def reset_transaksi(self):
        self.nama_barang.clear()
        self.harga_barang.clear()
        self.jumlah_barang.clear()
        print("Semua barang berhasil dihapus!")

# Memeriksa pesanan dan output transaksi barang
    def check_order(self):
        data = []
        for i in range(len(self.nama_barang)):

            data.append([i+1, self.nama_barang[i], self.jumlah_barang[i], self.harga_barang[i], self.harga_barang[i]*self.jumlah_barang[i]])
        print(tabulate(data, headers=["No.", "Nama Barang", "Jumlah", "Harga", "Total Harga"], tablefmt="fancy_grid"))
        return True

#memeriksa apakah dapat diskon
    def checkout(self):
        total_harga = 0
        for i in range(len(self.nama_barang)):
            total_harga += self.harga_barang[i] * self.jumlah_barang[i]

        if total_harga >= 500_000:
            total_harga *= 0.90
            print('Anda mendapatkan diskon 10% !')
        elif total_harga >= 300_000:
            total_harga *= 0.92
            print('Anda mendapatkan diskon 8% !')
        elif total_harga >= 200_000:
            total_harga *= 0.95
            print('Anda mendapatkan diskon 5% !')
        else:
            print('Anda tidak mendapatkan diskon')

        return total_harga

# Proses input interaktif

transaksi = Transaksi()

id_transaksi = int(input('Masukkan nomor ID Anda: '))
message = f'Welcome to Super Cashier {id_transaksi}'

print('*' * len(message))
print(message)
print('*' * len(message))

confirm1 = input('Apakah Anda akan berbelanja? (y/n) : ')
while confirm1.lower() == "y":
    print('\nKetik 1 untuk menambah item.')
    print('Ketik 2 untuk menghapus item.')
    print('Ketik 3 untuk mengedit item.')
    print('Ketik 0 untuk membatalkan belanja.')
    belanja = input('Masukkan pilihan Anda: ')
    if belanja == "1":
        nama_barang = input('Masukkan nama item: ')
        jumlah_barang = int(input('Masukkan jumlah item: '))
        harga_barang = float(input('Masukkan harga item: '))
        transaksi.tambah_barang(nama_barang, harga_barang, jumlah_barang)
        print("Barang berhasil ditambahkan.")
    elif belanja == "2":
        if len(transaksi.nama_barang) > 0:
            index_hapus = int(input('Masukkan nomor item yang ingin dihapus: '))
            if 1 <= index_hapus <= len(transaksi.nama_barang):
                index_hapus -= 1
                transaksi.hapus_barang(index_hapus)
                print('Item berhasil dihapus.')
            else:
                print('Nomor item tidak valid.')
        else:
            print('Tidak ada item dalam daftar belanja.')
    elif belanja == "3":
        if len(transaksi.nama_barang) > 0:
            index_edit = int(input('Masukkan nomor item yang ingin diedit: '))
            if 1 <= index_edit <= len(transaksi.nama_barang):
                index_edit -= 1
                nama_barang = input('Masukkan nama item baru: ')
                jumlah_barang = int(input('Masukkan jumlah item baru: '))
                harga_barang = float(input('Masukkan harga item baru: '))
                transaksi.update_barang(index_edit, nama_barang, harga_barang, jumlah_barang)
                print('Item berhasil diupdate.')
            else:
                print('Nomor item tidak valid.')
        else:
            print('Tidak ada item dalam daftar belanja.')
    elif belanja == "0":
        transaksi.reset_transaksi()
        print("Pemesanan berhasil dihapus.")
        break
    else:
        print('Pilihan tidak valid.')

    confirm1 = input('Apakah Anda ingin berbelanja lagi? (y/n) : ')
    if confirm1.lower() != "y":
        break

print('\nDaftar Belanja:')
if transaksi.check_order():
    total_harga = transaksi.checkout()
    print('Total Harga =', total_harga)
    print('Pemesanan sudah benar.')
else:
    print('Terjadi kesalahan input.')

print('\nTerima kasih atas kunjungan Anda')
```
